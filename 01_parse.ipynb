## bulk

```python
from_st = 'Kyiv'
to_st = 'Lviv'
```

```python
from multiprocessing.pool import ThreadPool

def f(iterable):
    date, from_st, to_st = iterable
    res = P.search_wagon_seats_by_date(date, from_st, to_st, num_try=10)
    return res
```

```python
main_dict = {}

for _ in tn(range(10_000)):
    iterable = [(pd.datetime.now() + pd.DateOffset(i), from_st, to_st) for i in range(90)]
    iterable += [(pd.datetime.now() + pd.DateOffset(i), to_st, from_st) for i in range(90)]
    iterable = sum([iterable[i::45] for i in range(45)], [])
    
    TP = ThreadPool(20)
    try:
        res = TP.map(f, iterable, chunksize=1)
        now = pd.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        main_dict[now] = res
        print(f'{now} - {sum(len(_) for _ in res)}')
    except Exception as e:
        now = pd.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        m = f'[uz_parse] error - {e}'
        print(m)
        tgn(m)
```
